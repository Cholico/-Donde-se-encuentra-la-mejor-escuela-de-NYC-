Cada año, los estudiantes de secundaria estadounidenses toman exámenes SAT, que son pruebas estandarizadas destinadas a medir la alfabetización, la aritmética y las habilidades de escritura. Hay tres secciones: lectura, matemáticas y escritura, cada una con una puntuación máxima de 800 puntos. Estas pruebas son extremadamente importantes para los estudiantes y las universidades, ya que juegan un papel fundamental en el proceso de admisión.

Analizar el desempeño de las escuelas es importante para una variedad de partes interesadas, incluidos los profesionales de la política y la educación, los investigadores, el gobierno e incluso los padres que consideran a qué escuela deben asistir sus hijos. 

Se le ha proporcionado un conjunto de datos llamado 'schools.csv', que se muestra a continuación.

Se le ha encomendado la tarea de responder tres preguntas clave sobre el desempeño del SAT en las escuelas públicas de la ciudad de Nueva York (NYC).



```python
# Re-run this cell 
import pandas as pd
import numpy as np

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

```python
best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)

best_math_schools.head(10)
```

```python
schools['total_SAT'] = schools["average_math"] + schools["average_reading"] + schools["average_writing"] 

schools_name = schools.groupby('school_name')['total_SAT'].mean().reset_index()

schools_name.sort_values(by="total_SAT", ascending=False)

schools_name.head(10)
```

```python
boroughs = schools.groupby('borough')['total_SAT'].agg(['count', 'mean', 'std']).round(2)

boroughs.head(10)
```

```python
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]


largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})

largest_std_dev
```
